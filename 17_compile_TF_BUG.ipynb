# Compare classification training of Keras model to TF model

0. Common

```python
import keras
import random
random.seed(1)

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_train = x_train.astype('float32') / 255.
y_train = keras.utils.to_categorical(y_train)

x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
x_test = x_test.astype('float32') / 255.
y_test = keras.utils.to_categorical(y_test)
```

1. Keras

```python
import keras
import sys; sys.path.insert(0, '..')
import models

model = keras.Sequential([
    models.build_encoder(16,1),
    keras.layers.Dense(10, activation='softmax')
])
# model = keras.Sequential([
#     encoder,
#     classifier
# ])


model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

encoder = model.get_layer('encoder1')
encoder.load_weights('../weights/CE16_epoch25_weights.h5')
for layer in encoder.layers:
    layer.trainable = False
# model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
# model.build(input_shape=(32,28,28,1))
print(model.summary())
print(encoder.weights[:1])

history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, batch_size=64)

print(encoder.weights[:1])
```
